```python
Suez_Canal_Daily_Transit_Ais['date'] = pd.to_datetime(Suez_Canal_Daily_Transit_Ais['date'], utc=True)
```

```python
individual_n_cols = ['n_container', 'n_dry_bulk', 'n_general_cargo', 'n_roro', 'n_tanker']
Suez_Canal_Daily_Transit_Ais['n_total_calculated'] = Suez_Canal_Daily_Transit_Ais[individual_n_cols].sum(axis=1)
```

```python
n_total_discrepancies = (Suez_Canal_Daily_Transit_Ais['n_total'] != Suez_Canal_Daily_Transit_Ais['n_total_calculated']).sum()
print(f"Number of n_total discrepancies: {n_total_discrepancies}")
```

```python
duplicates = Suez_Canal_Daily_Transit_Ais.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
print("\nPotential outliers (values beyond 3 standard deviations):")
for col in ['n_total', 'capacity']:
    mean = Suez_Canal_Daily_Transit_Ais[col].mean()
    std = Suez_Canal_Daily_Transit_Ais[col].std()
    outliers = Suez_Canal_Daily_Transit_Ais[(Suez_Canal_Daily_Transit_Ais[col] < mean - 3*std) | (Suez_Canal_Daily_Transit_Ais[col] > mean + 3*std)]
    print(f"{col}: {len(outliers)} potential outliers")
```

```python
# Calculate mean and standard deviation for our key metrics
n_total_mean = Suez_Canal_Daily_Transit_Ais['n_total'].mean()
n_total_std = Suez_Canal_Daily_Transit_Ais['n_total'].std()
cap_mean = Suez_Canal_Daily_Transit_Ais['capacity'].mean()
cap_std = Suez_Canal_Daily_Transit_Ais['capacity'].std()

# Define a function to find outliers
def find_outliers(series, mean, std):
    """Returns a DataFrame filter for values beyond 3 standard deviations."""
    return (series < mean - 3*std) | (series > mean + 3*std)

# Find the outlier rows
n_total_outliers = Suez_Canal_Daily_Transit_Ais[find_outliers(Suez_Canal_Daily_Transit_Ais['n_total'], n_total_mean, n_total_std)]
capacity_outliers = Suez_Canal_Daily_Transit_Ais[find_outliers(Suez_Canal_Daily_Transit_Ais['capacity'], cap_mean, cap_std)]

# Combine and deduplicate to see all unique days with any outlier
all_outliers = pd.concat([n_total_outliers, capacity_outliers]).drop_duplicates()
all_outliers = all_outliers.sort_values(by='date') # Sort chronologically

# Display the outliers in a clean format focusing on date and the key metrics
print("Dates with exceptional traffic (outliers):")
print(all_outliers[['date', 'n_total', 'capacity']].to_string(index=False))
```

```python
Suez_Canal_Daily_Transit_Ais = Suez_Canal_Daily_Transit_Ais.sort_values('date').reset_index(drop=True)
Suez_Canal_Daily_Transit_Ais = Suez_Canal_Daily_Transit_Ais.set_index('date')
```

```python
individual_n_cols = ['n_container','n_dry_bulk','n_general_cargo','n_roro','n_tanker']
Suez_Canal_Daily_Transit_Ais['n_total_calculated'] = Suez_Canal_Daily_Transit_Ais[individual_n_cols].sum(axis=1)
mismatches = Suez_Canal_Daily_Transit_Ais[Suez_Canal_Daily_Transit_Ais['n_total'] != Suez_Canal_Daily_Transit_Ais['n_total_calculated']]
print("n_total mismatch count:", mismatches.shape[0])
display(mismatches[['n_total','n_total_calculated']].head(20))
```

```python
# 4) preserve original n_total and (optionally) overwrite with calculated values
Suez_Canal_Daily_Transit_Ais['n_total_original'] = Suez_Canal_Daily_Transit_Ais.get('n_total_original')  # no-op if exists
Suez_Canal_Daily_Transit_Ais.loc[mismatches.index, 'n_total_original'] = Suez_Canal_Daily_Transit_Ais.loc[mismatches.index, 'n_total']
# Overwrite (recommended if per-type counts are authoritative):
Suez_Canal_Daily_Transit_Ais.loc[mismatches.index, 'n_total'] = Suez_Canal_Daily_Transit_Ais.loc[mismatches.index, 'n_total_calculated']
```

```python
# 5) inspect the suspicious 2023-11-17 spike (and any outliers chronologically)
print("Row(s) for 2023-11-17:")
display(Suez_Canal_Daily_Transit_Ais.loc['2023-11-17': '2023-11-17'])
```

```python
n_mean, n_std = Suez_Canal_Daily_Transit_Ais['n_total'].mean(), Suez_Canal_Daily_Transit_Ais['n_total'].std()
cap_mean, cap_std = Suez_Canal_Daily_Transit_Ais['capacity'].mean(), Suez_Canal_Daily_Transit_Ais['capacity'].std()
Suez_Canal_Daily_Transit_Ais['is_outlier'] = (
    (Suez_Canal_Daily_Transit_Ais['n_total'] < n_mean - 3*n_std) |
    (Suez_Canal_Daily_Transit_Ais['n_total'] > n_mean + 3*n_std) |
    (Suez_Canal_Daily_Transit_Ais['capacity'] < cap_mean - 3*cap_std) |
    (Suez_Canal_Daily_Transit_Ais['capacity'] > cap_mean + 3*cap_std))
```

```python
if not isinstance(Babel_el_Mande, pd.DataFrame):
    raise TypeError(f"Babel_el_Mande must be a pandas DataFrame, got {type(Babel_el_Mande)}")

# Ensure a proper datetime 'date' column exists
if 'date' not in Babel_el_Mande.columns or Babel_el_Mande['date'].isna().any():
    if set(['year','month','day']).issubset(Babel_el_Mande.columns):
        Babel_el_Mande['date'] = pd.to_datetime(
            Babel_el_Mande[['year','month','day']].assign(hour=0),
            errors='coerce', utc=True
        )
    else:
        # try parsing existing date column if present
        if 'date' in Babel_el_Mande.columns:
            Babel_el_Mande['date'] = pd.to_datetime(Babel_el_Mande['date'], errors='coerce', utc=True)
        else:
            raise RuntimeError("No 'date' column and no year/month/day to build it from.")

# Small helper: robust outlier flagger (doesn't drop)
def flag_outliers_inplace(df, cols=['n_total','capacity'], z_thresh=3.0, mad_thresh=3.5):
    """
    Adds flags to df (inplace) for outliers using three methods:
      - 'out_z_<col>'   : classic z-score (mean/std)
      - 'out_mad_<col>' : robust modified z by MAD
      - 'out_logz_<col>': z-score on log1p (for heavy tail; only non-negative)
      - 'out_any_<col>' : combined OR of above
    Returns: dictionary with counts and a DataFrame of outlier rows for each column.
    """
    results = {}
    for col in cols:
        if col not in df.columns:
            results[col] = {"count": 0, "outliers_df": pd.DataFrame()}
            continue

        s = pd.to_numeric(df[col], errors='coerce')

        # z-score (mean/std)
        mean = s.dropna().mean()
        std = s.dropna().std()
        if std == 0 or np.isnan(std):
            out_z = pd.Series(False, index=df.index)
        else:
            out_z = (s < (mean - z_thresh * std)) | (s > (mean + z_thresh * std))

        # MAD (median absolute deviation -> modified z)
        med = s.dropna().median()
        mad = (s.dropna() - med).abs().median()
        if mad == 0 or np.isnan(mad):
            out_mad = pd.Series(False, index=df.index)
        else:
            mod_z = 0.6745 * (s - med) / mad
            out_mad = mod_z.abs() > mad_thresh

        # log-based z-score for non-negative values (useful for capacity)
        log_flag = pd.Series(False, index=df.index)
        mask_nonneg = s.notna() & (s >= 0)
        if mask_nonneg.sum() > 1:
            log_vals = np.log1p(s[mask_nonneg])
            log_mean = log_vals.mean()
            log_std = log_vals.std()
            if log_std not in (0, np.nan):
                tmp = (log_vals < (log_mean - z_thresh * log_std)) | (log_vals > (log_mean + z_thresh * log_std))
                log_flag.loc[mask_nonneg] = tmp

        # Combined
        out_any = out_z | out_mad | log_flag.fillna(False)

        # Write flags into df (inplace)
        df[f'out_z_{col}'] = out_z.fillna(False)
        df[f'out_mad_{col}'] = out_mad.fillna(False)
        df[f'out_logz_{col}'] = log_flag.fillna(False)
        df[f'out_any_{col}'] = out_any.fillna(False)

        # Build DataFrame of outliers with date and useful context (sorted desc by value)
        outliers_df = df.loc[out_any, ['date', 'portname'] + [col, f'out_z_{col}', f'out_mad_{col}', f'out_logz_{col}']].copy()
        outliers_df = outliers_df.sort_values(by=col, ascending=False).reset_index(drop=True)

        results[col] = {
            "count": int(out_any.sum()),
            "outliers_df": outliers_df,
            "stats": {"mean": mean, "std": std, "median": med, "mad": mad}
        }

    return results

# Run the flagger on your DataFrame
results = flag_outliers_inplace(Babel_el_Mande, cols=['n_total', 'capacity'], z_thresh=3.0, mad_thresh=3.5)

# Print counts and show example dates for each column
for col, info in results.items():
    print(f"\nColumn: {col}")
    print(f"Outlier count (any method): {info['count']}")
    stats = info['stats']
    print(f"Stats: mean={stats['mean']:.3g}, std={stats['std']:.3g}, median={stats['median']:.3g}, mad={stats['mad']:.3g}")
    if info['count'] > 0:
        # show top 20 outlier rows (date, port, value, flags)
        display(info['outliers_df'].head(20))
    else:
        print("No outliers flagged for this column.")
```

```python
# === LOAD CLEANED DATASETS ===
chokepoints = {
    'Cape_of_Good_Hope': Clean_Cape_of_Good_Hope,
    'Suez_Canal': Clean_Suez_Canal,
    'Panama_Canal': Clean_Panama_Canal,
    'Strait_Of_Hormuz': Clean_Strait_Of_Hormuz,
    'Babel_el_Mandeb': Clean_Babel_el_Mande,
    'Strait_of_Malacca': Clean_Strait_of_Malacca
}

print("=== MARITIME CHOKEPOINTS ANALYSIS ===")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set up plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# 1. MARITIME CHOKEPOINTS ANALYSIS
print("=== MARITIME CHOKEPOINTS ANALYSIS ===\n")

# List of chokepoint datasets
chokepoints = {
    'Cape_of_Good_Hope': Clean_Cape_of_Good_Hope,
    'Suez_Canal': Clean_Suez_Canal,
    'Panama_Canal': Clean_Panama_Canal,
    'Strait_Of_Hormuz': Clean_Strait_Of_Hormuz,
    'Babel_el_Mandeb': Clean_Babel_el_Mande,
    'Strait_of_Malacca': Clean_Strait_of_Malacca
}

# Basic analysis for each chokepoint
for name, df in chokepoints.items():
    # compute simple time period info safely
    if 'date' in df.columns:
        start = df['date'].min()
        end = df['date'].max()
    else:
        # fall back to year-month-day columns if present
        if all(col in df.columns for col in ['year', 'month', 'day']):
            start = pd.to_datetime(df.loc[df.index.min(), ['year','month','day']])
            end = pd.to_datetime(df.loc[df.index.max(), ['year','month','day']])
        else:
            start = end = None

    print(f"\n--- {name.replace('_', ' ')} ---")
    print(f"Time period: {start} to {end}")
    print(f"Total records: {len(df):,}")
    print(f"Unique ports: {df['portname'].nunique()}")
    print(f"Average daily traffic: {df['n_total'].mean():.0f} ships")
    # capacity might not exist in all frames
    if 'capacity' in df.columns:
        print(f"Average daily capacity: {df['capacity'].mean():,.0f} tons")
    else:
        print("Average daily capacity: N/A")
    print(f"Outlier percentage: {df['is_outlier'].mean()*100:.1f}%" if 'is_outlier' in df.columns else "No outlier info")

# 2. VISUALIZATION: TRAFFIC COMPARISON ACROSS CHOKEPOINTS
fig, axes = plt.subplots(2, 2, figsize=(20, 15))
fig.suptitle('Maritime Chokepoints Traffic Analysis', fontsize=16, fontweight='bold')

# Build monthly averages for each chokepoint (ensure a consistent 'date' column for month start)
monthly_traffic = []
for name, df in chokepoints.items():
    # If df has a proper datetime 'date' column -> use resample (safer & cleaner)
    if 'date' in df.columns and pd.api.types.is_datetime64_any_dtype(df['date']):
        monthly = df.set_index('date').resample('M')['n_total'].mean().reset_index()
        # Convert to month-start (1st of month) for plotting consistency
        monthly['date'] = monthly['date'].dt.to_period('M').dt.to_timestamp()
    else:
        # fallback to grouping by year & month columns
        if 'year' in df.columns and 'month' in df.columns:
            monthly = df.groupby(['year', 'month'])['n_total'].mean().reset_index()
            monthly['date'] = pd.to_datetime(monthly[['year', 'month']].assign(day=1))
        else:
            # If neither available, skip this chokepoint
            continue

    monthly = monthly[['date', 'n_total']].copy()
    monthly['chokepoint'] = name
    monthly_traffic.append(monthly)

# Concatenate safely (handles the case where some chokepoints were skipped)
if monthly_traffic:
    monthly_df = pd.concat(monthly_traffic, ignore_index=True)
else:
    monthly_df = pd.DataFrame(columns=['date', 'n_total', 'chokepoint'])

sns.lineplot(data=monthly_df, x='date', y='n_total', hue='chokepoint', ax=axes[0,0])
axes[0,0].set_title('Monthly Average Ship Traffic')
axes[0,0].set_xlabel('Date')
axes[0,0].set_ylabel('Number of Ships')
axes[0,0].tick_params(axis='x', rotation=45)

# Plot 2: Ship type distribution
ship_types = ['n_container', 'n_dry_bulk', 'n_tanker', 'n_roro', 'n_general_cargo']
ship_type_avg = []
for name, df in chokepoints.items():
    for st in ship_types:
        if st in df.columns:
            ship_type_avg.append({
                'chokepoint': name,
                'ship_type': st.replace('n_', '').replace('_', ' ').title(),
                'avg_count': df[st].mean()
            })

ship_type_df = pd.DataFrame(ship_type_avg)
if not ship_type_df.empty:
    pivot_df = ship_type_df.pivot(index='chokepoint', columns='ship_type', values='avg_count').fillna(0)
    pivot_df.plot(kind='bar', ax=axes[0,1], stacked=True)
    axes[0,1].set_title('Average Ship Type Distribution by Chokepoint')
    axes[0,1].set_xlabel('Chokepoint')
    axes[0,1].set_ylabel('Average Number of Ships')
    axes[0,1].tick_params(axis='x', rotation=45)
    axes[0,1].legend(title='Ship Type')

# Plot 3: Capacity utilization
capacity_data = []
for name, df in chokepoints.items():
    if 'capacity' in df.columns and 'n_total' in df.columns:
        capacity_data.append({
            'chokepoint': name,
            'avg_capacity': df['capacity'].mean(),
            'avg_ships': df['n_total'].mean()
        })

capacity_df = pd.DataFrame(capacity_data)
if not capacity_df.empty:
    axes[1,0].scatter(capacity_df['avg_ships'], capacity_df['avg_capacity'], s=100)
    for i, row in capacity_df.iterrows():
        axes[1,0].annotate(row['chokepoint'].replace('_', '\n'),
                          (row['avg_ships'], row['avg_capacity']),
                          xytext=(5, 5), textcoords='offset points')
    axes[1,0].set_xlabel('Average Number of Ships')
    axes[1,0].set_ylabel('Average Capacity (tons)')
    axes[1,0].set_title('Capacity vs Ship Count by Chokepoint')

# Plot 4: Seasonal patterns
seasonal_data = []
for name, df in chokepoints.items():
    if 'month' in df.columns:
        seasonal = df.groupby('month')['n_total'].mean().reset_index()
        seasonal['chokepoint'] = name
        seasonal_data.append(seasonal)

if seasonal_data:
    seasonal_df = pd.concat(seasonal_data, ignore_index=True)
    sns.lineplot(data=seasonal_df, x='month', y='n_total', hue='chokepoint', ax=axes[1,1])
    axes[1,1].set_title('Seasonal Traffic Patterns')
    axes[1,1].set_xlabel('Month')
    axes[1,1].set_ylabel('Average Number of Ships')
    axes[1,1].set_xticks(range(1, 13))
else:
    axes[1,1].text(0.5, 0.5, 'No month data available', ha='center')

plt.tight_layout()
plt.show()

# 3. RUSSIAN OIL ANALYSIS
print("\n=== RUSSIAN FOSSIL FUEL EXPORTS ANALYSIS ===\n")

# Russian Oil Import analysis
print("Top importers of Russian fossil fuels:")
russian_import_summary = Russian_Oil_Import.groupby('destination_country')[['Oil', 'Coal', 'Gas']].sum().sort_values('Oil', ascending=False)
print(russian_import_summary.head(10))

# Visualization for Russian oil data
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Russian Fossil Fuel Export Analysis', fontsize=16, fontweight='bold')

# Plot 1: Top importers by fuel type
top_countries = Russian_Oil_Import.groupby('destination_country')[['Oil', 'Coal', 'Gas']].sum().sum(axis=1).nlargest(10).index
top_data = Russian_Oil_Import[Russian_Oil_Import['destination_country'].isin(top_countries)]
melted_data = top_data.melt(id_vars=['destination_country'],
                           value_vars=['Oil', 'Coal', 'Gas'],
                           var_name='Fuel_Type',
                           value_name='Value')

sns.barplot(data=melted_data, x='destination_country', y='Value', hue='Fuel_Type', ax=axes[0,0])
axes[0,0].set_title('Top Importers of Russian Fossil Fuels by Type')
axes[0,0].set_xlabel('Country')
axes[0,0].set_ylabel('Total Volume')
axes[0,0].tick_params(axis='x', rotation=45)

# Plot 2: Regional analysis
regional_data = Russian_Oil_Import.groupby('destination_region')[['Oil', 'Coal', 'Gas']].sum()
regional_data.plot(kind='bar', ax=axes[0,1])
axes[0,1].set_title('Russian Fossil Fuel Imports by Region')
axes[0,1].set_xlabel('Region')
axes[0,1].set_ylabel('Total Volume')
axes[0,1].tick_params(axis='x', rotation=45)

# Plot 3: Russian Oil Exports time series
Russian_Oil_Exports['date'] = pd.to_datetime(Russian_Oil_Exports['date'], errors='coerce')
exports_ts = Russian_Oil_Exports.groupby('date')[['Crude oil', 'Oil products and chemicals',
                                                'Pipeline gas', 'Coal', 'LNG']].sum()

exports_ts.plot(ax=axes[1,0])
axes[1,0].set_title('Russian Fossil Fuel Export Value Over Time')
axes[1,0].set_xlabel('Date')
axes[1,0].set_ylabel('Export Value (EUR)')
axes[1,0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot 4: Export composition
export_composition = Russian_Oil_Exports[['Crude oil', 'Oil products and chemicals',
                                        'Pipeline gas', 'Coal', 'LNG']].sum()
axes[1,1].pie(export_composition, labels=export_composition.index, autopct='%1.1f%%', startangle=90)
axes[1,1].set_title('Composition of Russian Fossil Fuel Exports')

plt.tight_layout()
plt.show()

# 4. CORRELATION ANALYSIS BETWEEN CHOKEPOINTS AND RUSSIAN OIL
print("\n=== CORRELATION ANALYSIS ===\n")

# Create monthly aggregates for correlation analysis
if not monthly_df.empty:
    monthly_traffic_all = monthly_df.pivot(index='date', columns='chokepoint', values='n_total').fillna(0)
else:
    monthly_traffic_all = pd.DataFrame()

# Align Russian oil data with traffic data
if 'date' in Russian_Oil_Exports.columns:
    russian_monthly = Russian_Oil_Exports.groupby(pd.Grouper(key='date', freq='M'))[
        ['Crude oil', 'Oil products and chemicals', 'Pipeline gas', 'Coal', 'LNG']].sum()
else:
    russian_monthly = pd.DataFrame()

# Merge datasets for correlation analysis
if not monthly_traffic_all.empty and not russian_monthly.empty:
    correlation_data = monthly_traffic_all.join(russian_monthly, how='inner')

    if not correlation_data.empty:
        # Calculate correlations
        correlation_matrix = correlation_data.corr()

        # Plot correlation heatmap
        plt.figure(figsize=(12, 10))
        mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
        sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
                    square=True, fmt='.2f')
        plt.title('Correlation: Chokepoint Traffic vs Russian Fossil Fuel Exports')
        plt.tight_layout()
        plt.show()

        # Print strongest correlations
        oil_corrs = correlation_matrix['Crude oil'].sort_values(ascending=False)
        print("Strongest correlations with Crude oil exports:")
        print(oil_corrs.head(6))

        products_corrs = correlation_matrix['Oil products and chemicals'].sort_values(ascending=False)
        print("\nStrongest correlations with Oil products exports:")
        print(products_corrs.head(6))

# 5. OUTLIER ANALYSIS
print("\n=== OUTLIER ANALYSIS ===\n")

for name, df in chokepoints.items():
    if 'is_outlier' in df.columns:
        outliers = df[df['is_outlier'] == True]
        if len(outliers) > 0:
            print(f"\n{name}: {len(outliers)} outliers detected")
            print(f"Outlier dates: {outliers['date'].tolist() if 'date' in outliers.columns else 'Check year-month-day columns'}")

# 6. TREND ANALYSIS
print("\n=== TREND ANALYSIS ===\n")

# Calculate trends for each chokepoint
trend_results = []
for name, df in chokepoints.items():
    if len(df) > 12:  # Only calculate trend if we have sufficient data
        # Create time variable
        if 'date' in df.columns and pd.api.types.is_datetime64_any_dtype(df['date']):
            df_sorted = df.sort_values('date').copy()
            time_var = (df_sorted['date'] - df_sorted['date'].min()).dt.days
        else:
            df_sorted = df.sort_values(['year', 'month', 'day']).copy()
            df_sorted['date'] = pd.to_datetime(df_sorted[['year', 'month', 'day']])
            time_var = (df_sorted['date'] - df_sorted['date'].min()).dt.days

        # Calculate linear trend
        z = np.polyfit(time_var, df_sorted['n_total'], 1)
        trend_slope = z[0]  # Ships per day

        trend_results.append({
            'chokepoint': name,
            'trend_slope': trend_slope,
            'avg_daily_traffic': df_sorted['n_total'].mean(),
            'trend_percentage': (trend_slope * 365) / df_sorted['n_total'].mean() * 100
        })

trend_df = pd.DataFrame(trend_results)
if not trend_df.empty:
    print("Annual traffic trends:")
    for _, row in trend_df.iterrows():
        direction = "increasing" if row['trend_slope'] > 0 else "decreasing"
        print(f"{row['chokepoint']}: {direction} by {abs(row['trend_percentage']):.1f}% per year")

# 7. ADDITIONAL INSIGHTS
print("\n=== KEY INSIGHTS ===\n")

if not monthly_traffic_all.empty:
    busiest = monthly_traffic_all.mean().idxmax()
    print(f"Busiest chokepoint: {busiest} (avg {monthly_traffic_all[busiest].mean():.0f} ships/month)")

    volatility = monthly_traffic_all.std() / monthly_traffic_all.mean()
    most_volatile = volatility.idxmax()
    print(f"Most volatile chokepoint: {most_volatile} (coefficient of variation: {volatility[most_volatile]:.2f})")
else:
    print("Not enough monthly chokepoint data to compute busiest/volatility.")

# Russian oil export trend (if data exists)
if 'date' in Russian_Oil_Exports.columns and not russian_monthly.empty:
    oil_trend = russian_monthly['Crude oil'].pct_change(12).mean() * 100
    print(f"Average annual crude oil export trend: {oil_trend:.1f}%")

print("\n=== ANALYSIS COMPLETE ===")
```

```python
geopolitical_events = {
    '2019-05-15': 'US-China Trade War escalation',
    '2020-03-11': 'COVID-19 declared pandemic', 
    '2020-04-01': 'Global lockdowns begin',
    '2021-03-23': 'Suez Canal blockage (Ever Given)',
    '2022-02-24': 'Russia-Ukraine War begins',
    '2022-06-05': 'Russian oil sanctions implemented',
    '2023-07-01': 'Panama Canal drought restrictions',
    '2023-10-07': 'Red Sea attacks begin',
    '2024-01-12': 'US/UK strikes in Yemen'
}

print("Available datasets for correlation:")
print("- Maritime traffic data (2019-2025)")
print("- Nifty 50 stock index")
print("- Nasdaq index") 
print("- Russian oil import/export data")
print("- WTI Oil Price data")

# Fixed function to analyze stock market reactions to maritime disruptions
def analyze_event_impact(df, event_date, event_name, window_days=30, price_col='Close'):
    """
    Analyze financial market impact around geopolitical events
    
    Parameters:
    - df: DataFrame with 'Date' column
    - event_date: Date string in 'YYYY-MM-DD' format
    - event_name: Name of the event
    - window_days: Number of days before/after event to analyze
    - price_col: Column name to analyze (default 'Close')
    """
    
    # Convert event_date to datetime
    event_dt = pd.to_datetime(event_date)
    
    # Make sure Date column is datetime
    if df['Date'].dtype != 'datetime64[ns]':
        df['Date'] = pd.to_datetime(df['Date'])
    
    # Sort by date to ensure proper ordering
    df_sorted = df.sort_values('Date').copy()
    
    # Find pre-event and post-event windows
    pre_window = df_sorted[df_sorted['Date'] < event_dt].tail(window_days)
    post_window = df_sorted[df_sorted['Date'] >= event_dt].head(window_days)
    
    # Check if we have enough data
    if len(pre_window) == 0 or len(post_window) == 0:
        return {
            'event': event_name,
            'event_date': event_date,
            'status': 'Insufficient data',
            'pre_event_avg': np.nan,
            'post_event_avg': np.nan, 
            'impact_percentage': np.nan,
            'volatility_change': np.nan
        }
    
    # Calculate impact metrics for the price column
    pre_avg = pre_window[price_col].mean()
    post_avg = post_window[price_col].mean()
    impact_pct = ((post_avg - pre_avg) / pre_avg) * 100
    
    # Calculate volatility change
    pre_volatility = pre_window[price_col].std()
    post_volatility = post_window[price_col].std()
    volatility_change = post_volatility / pre_volatility if pre_volatility > 0 else np.nan
    
    # Calculate return metrics if Return column exists
    return_metrics = {}
    if 'Return' in df_sorted.columns:
        pre_return_avg = pre_window['Return'].mean() if len(pre_window) > 0 else np.nan
        post_return_avg = post_window['Return'].mean() if len(post_window) > 0 else np.nan
        return_metrics = {
            'pre_return_avg': pre_return_avg,
            'post_return_avg': post_return_avg,
            'return_difference': post_return_avg - pre_return_avg
        }
    
    result = {
        'event': event_name,
        'event_date': event_date,
        'status': 'Success',
        'pre_event_avg': pre_avg,
        'post_event_avg': post_avg, 
        'impact_percentage': impact_pct,
        'volatility_change': volatility_change,
        'pre_days_analyzed': len(pre_window),
        'post_days_analyzed': len(post_window)
    }
    
    # Add return metrics if they exist
    result.update(return_metrics)
    
    return result

# Function to analyze all events for a given dataset
def analyze_all_events(df, dataset_name, window_days=30, price_col='Close'):
    """Analyze all geopolitical events for a given dataset"""
    
    print(f"\n{'='*60}")
    print(f"Analyzing {dataset_name}")
    print(f"{'='*60}")
    
    event_impacts = []
    for event_date, event_name in geopolitical_events.items():
        impact = analyze_event_impact(df, event_date, event_name, window_days, price_col)
        event_impacts.append(impact)
        
        # Print individual event results
        if impact['status'] == 'Success':
            print(f"\n{event_name} ({event_date}):")
            print(f"  Price Impact: {impact['impact_percentage']:.2f}%")
            print(f"  Volatility Change: {impact['volatility_change']:.2f}x")
            if 'return_difference' in impact:
                print(f"  Return Change: {impact['return_difference']:.4f}")
    
    # Create DataFrame for easier analysis
    impacts_df = pd.DataFrame(event_impacts)
    impacts_df = impacts_df[impacts_df['status'] == 'Success']  # Filter successful analyses
    
    return impacts_df

# Apply to Nifty data
print("\nAnalyzing Nifty 500 Index...")
nifty_impacts = analyze_all_events(Nifty_data, "Nifty 500", window_days=30, price_col='Close')

# Apply to Nasdaq data
print("\nAnalyzing Nasdaq Index...")
nasdaq_impacts = analyze_all_events(Nasdaq, "Nasdaq", window_days=30, price_col='Close/Last')

# Summary statistics
print("\n" + "="*60)
print("SUMMARY STATISTICS")
print("="*60)

print("\nNifty 500 - Average Impact Across All Events:")
if len(nifty_impacts) > 0:
    print(f"  Average Price Impact: {nifty_impacts['impact_percentage'].mean():.2f}%")
    print(f"  Average Volatility Change: {nifty_impacts['volatility_change'].mean():.2f}x")
    print(f"  Most Impactful Event: {nifty_impacts.loc[nifty_impacts['impact_percentage'].abs().idxmax(), 'event']}")
    print(f"    Impact: {nifty_impacts.loc[nifty_impacts['impact_percentage'].abs().idxmax(), 'impact_percentage']:.2f}%")

print("\nNasdaq - Average Impact Across All Events:")
if len(nasdaq_impacts) > 0:
    print(f"  Average Price Impact: {nasdaq_impacts['impact_percentage'].mean():.2f}%")
    print(f"  Average Volatility Change: {nasdaq_impacts['volatility_change'].mean():.2f}x")
    print(f"  Most Impactful Event: {nasdaq_impacts.loc[nasdaq_impacts['impact_percentage'].abs().idxmax(), 'event']}")
    print(f"    Impact: {nasdaq_impacts.loc[nasdaq_impacts['impact_percentage'].abs().idxmax(), 'impact_percentage']:.2f}%")
```

```python
def preprocess_unfied(unfied):
    """Clean types, parse dates, create useful columns, and coerce numeric columns."""
    df = unfied.copy()

    # parse date
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    # drop rows with invalid dates
    df = df.dropna(subset=['date']).reset_index(drop=True)
    
    # ensure integer/numeric cargo columns
    cargo_cols = ['n_container','n_dry_bulk','n_general_cargo','n_roro','n_tanker','n_cargo','n_total']
    for c in cargo_cols:
        df[c] = pd.to_numeric(df[c], errors='coerce').fillna(0).astype(int)
    
    # capacity columns
    cap_cols = ['capacity_container','capacity_dry_bulk','capacity_general_cargo','capacity_roro','capacity_tanker','capacity_cargo','capacity']
    for c in cap_cols:
        df[c] = pd.to_numeric(df[c], errors='coerce')
    
    # booleanize outlier flags if present
    df['is_outlier'] = df['is_outlier'].astype(str).str.lower().map({'true': True, 'false': False}).fillna(False)
    
    # create month/year columns if not present
    if 'year' not in df.columns:
        df['year'] = df['date'].dt.year
    if 'month' not in df.columns:
        df['month'] = df['date'].dt.month
    if 'day' not in df.columns:
        df['day'] = df['date'].dt.day
    
    # cargo mix columns (fractions)
    df['container_frac'] = df['n_container'] / df['n_total'].replace(0, np.nan)
    df['tanker_frac']    = df['n_tanker'] / df['n_total'].replace(0, np.nan)
    df['drybulk_frac']   = df['n_dry_bulk'] / df['n_total'].replace(0, np.nan)
    df['general_frac']   = df['n_general_cargo'] / df['n_total'].replace(0, np.nan)
    df['roro_frac']      = df['n_roro'] / df['n_total'].replace(0, np.nan)
    df['cargo_frac']     = df['n_cargo'] / df['n_total'].replace(0, np.nan)

    # capacity utilization (observed / capacity)
    df['util_total'] = df['n_total'] / df['capacity'].replace(0, np.nan)
    
    return df

# usage
unified = preprocess_unfied(unfied_data)
```

```python
def cargo_summary_time(unified_df, freq='W'):
    """Return time-series of aggregated cargo counts and capacity utilization at a given freq (D/W/M)."""
    aggs = {
        'n_container':'sum','n_dry_bulk':'sum','n_general_cargo':'sum','n_roro':'sum',
        'n_tanker':'sum','n_cargo':'sum','n_total':'sum','capacity':'sum'
    }
    ts = unified_df.set_index('date').resample(freq).agg(aggs).fillna(0)
    ts['util_total'] = ts['n_total'] / ts['capacity'].replace(0, np.nan)
    return ts

def cargo_by_chokepoint(unified_df):
    """Aggregate cargo counts by chokepoint (date index)."""
    cols = ['n_container','n_dry_bulk','n_general_cargo','n_roro','n_tanker','n_cargo','n_total']
    df = unified_df.groupby(['chokepoint','date'])[cols].sum().reset_index()
    return df

def cargo_by_port(unified_df, top_n=20):
    """Top ports by total traffic."""
    port_totals = unified_df.groupby('portname')['n_total'].sum().sort_values(ascending=False)
    top_ports = port_totals.head(top_n).index.tolist()
    top_df = unified_df[unified_df['portname'].isin(top_ports)].copy()
    summary = top_df.groupby('portname')[['n_total','n_container','n_tanker']].sum().sort_values('n_total', ascending=False)
    return summary

# usage
ts_weekly = cargo_summary_time(unified, freq='W')
chokepoint_df = cargo_by_chokepoint(unified)
top_ports = cargo_by_port(unified, top_n=15)
```

```python
def plot_cargo_mix(ts, start=None, end=None, ax=None):
    """Stacked area showing cargo type evolution over time (ts is resampled time series)."""
    data = ts.copy()
    if start: data = data.loc[start:]
    if end: data = data.loc[:end]
    mix = data[['n_container','n_tanker','n_dry_bulk','n_general_cargo','n_roro']]
    ax = ax or plt.gca()
    mix.plot.area(alpha=0.85, ax=ax)
    ax.set_title('Cargo Type Mix Over Time')
    ax.set_ylabel('Counts (weekly sum)')
    ax.legend(loc='upper left')
    plt.tight_layout()


# usage
plot_cargo_mix(ts_weekly)

```

```python
from scipy import stats

def chokepoint_util_stats(unified_df):
    """Return mean/std/utilization statistics by chokepoint."""
    cols = ['n_total','capacity','util_total']
    stats_df = unified_df.groupby('chokepoint')[cols].agg(['mean','std','median','count'])
    return stats_df

def flag_outliers_by_port(unified_df, z_thresh=3.0):
    """Flag days per port where n_total is an outlier using z-score on rolling window."""
    df = unified_df.copy()
    df = df.sort_values(['portname','date'])
    def mark(group):
        z = np.abs(stats.zscore(group['n_total'].replace(0, np.nan).fillna(method='ffill').fillna(0)))
        group['zscore_n_total'] = z
        group['is_outlier_n_total'] = z > z_thresh
        return group
    out = df.groupby('portname').apply(mark)
    return out

# usage
cp_stats = chokepoint_util_stats(unified)
unified_with_outliers = flag_outliers_by_port(unified, z_thresh=3.5)
```

```python
def event_window_summary(unified_df, event_date, window_days=30, chokepoint=None):
    """
    Summarize % change in traffic pre vs post event_date.
    Returns average daily traffic pre & post and % change.
    """
    # Ensure event date is timezone-aware to match your data
    ed = pd.to_datetime(event_date).tz_localize("UTC")  

    # Make sure 'date' column is datetime and tz-aware
    unified_df = unified_df.copy()
    if not pd.api.types.is_datetime64_any_dtype(unified_df['date']):
        unified_df['date'] = pd.to_datetime(unified_df['date'])
    if unified_df['date'].dt.tz is None:
        unified_df['date'] = unified_df['date'].dt.tz_localize("UTC")

    # Pre and post windows
    pre = unified_df[(unified_df['date'] >= (ed - pd.Timedelta(days=window_days))) & 
                     (unified_df['date'] < ed)]
    post = unified_df[(unified_df['date'] <= (ed + pd.Timedelta(days=window_days))) & 
                      (unified_df['date'] > ed)]

    if chokepoint:
        pre = pre[pre['chokepoint'] == chokepoint]
        post = post[post['chokepoint'] == chokepoint]

    pre_avg = pre['n_total'].mean()
    post_avg = post['n_total'].mean()
    pct_change = (post_avg - pre_avg) / pre_avg * 100 if pre_avg and not np.isnan(pre_avg) else np.nan

    return {
        'event_date': ed,
        'pre_avg': pre_avg,
        'post_avg': post_avg,
        'pct_change': pct_change
    }
```
